```python
## Dependencies
import os
import pandas as pd
```

## Suicide Rates

```python
# Set CSV input file path
input_path = os.path.join('input/owid_suicide-death-rates_per_100k.csv')
# Read CSV to DataFrame
df = pd.read_csv(input_path)
```

```python
df
```

```python
countries_list = pd.unique(df['Entity'])
len(countries_list)
```

```python
clean_df = df.rename(columns={
    'Entity': 'country',
    'Code': 'country_code',
    'Year': 'year',
    'Deaths - Self-harm - Sex: Both - Age: Age-standardized (Rate)': 'suicide_rate'
})
```

```python
clean_df.to_csv("clean/suicide_rate_1990.csv", index=True)
```
