```python
## Dependencies
import os
import numpy
import pandas as pd
```

## Suicide Rates

```python
# Set CSV input file path
suicide_path = os.path.join('input/owid_suicide-death-rates_per_100k.csv')
# Read CSV to DataFrame
suicide_df = pd.read_csv(suicide_path)
```

```python
suicide_df
```

```python
suicide_countries = pd.unique(suicide_df['Entity'])
len(suicide_countries)
```

```python
suicide_df.dtypes
```

```python
suicide_df = suicide_df.rename(columns={'Entity':'country','Code': 'country_code','Year': 'year','Deaths - Self-harm - Sex: Both - Age: Age-standardized (Rate)': 'suicide_rate_owid'})
```

```python
suicide = suicide_df[["country","year","suicide_rate_owid"]]
```

```python
suicide["year"]=suicide["year"].astype(str)
```

```python
suicide["country_year"]=suicide["country"]+"-"+suicide["year"]
```

```python
suicide
```

```python
suicide.to_csv("clean/extract_owid_suicide_rate_1990.csv", index=False)
```

## Life Expectancy

```python
# Set CSV input file path
life_path = os.path.join('input/owid_life-expectancy-at-birth-total-years.csv')
# Read CSV to DataFrame
life_df = pd.read_csv(life_path, delimiter=',', sep=r', ')
```

```python
life_df
```

```python
life_countries = pd.unique(life_df['Entity'])
len(life_countries)
```

```python
life_df = life_df.rename(columns={'Entity':'country','Code':'country_code','Year':'year','Life expectancy at birth, total (years)': 'life_expectancy'})
life_df
```

```python
life_expectancy = life_df[["country","year","life_expectancy"]]
```

```python
life_all =life_expectancy.loc[(life_expectancy['year'] >= 1990) & ((life_expectancy['year'] <= 2020))]
```

```python
life_countries_years = pd.unique(life_all['country'])
len(life_countries_years)
```

```python
life_all["year"]=life_all["year"].astype(str)
life_all["country_year"]=life_all["country"]+"-"+life_all["year"]
```

```python
life_all.to_csv("clean/extract_owid_life_expectancy.csv", index=False)
```

## Mental & Substance Disorders

```python
# Set CSV input file path
mental_path = os.path.join('input/owid_share-with-mental-and-substance-disorders_pct.csv')
# Read CSV to DataFrame
mental_df = pd.read_csv(mental_path, delimiter=',', sep=r', ')
mental_df
```

```python
mental_countries = pd.unique(mental_df['Entity'])
len(mental_countries)
```

```python
mental_ = mental_df[["country","year","mental_substance_disorders"]]
mental_
```

```python
mental_["year"]=mental_["year"].astype(str)
```

```python
mental_["country_year"]=mental_["country"]+"-"+mental_["year"]
```

```python
mental_
```

```python
mental_.to_csv("clean/extract_owid_mental_and_disorders.csv", index=False)
```
