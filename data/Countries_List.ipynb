```python
## Dependencies
import os
import pandas as pd
```

```python
# Set CSV input file path
input_path = os.path.join('country_list.csv')
# Read CSV to DataFrame
df = pd.read_csv(input_path, encoding = "ISO-8859-1")
df
```

```python
countries = df[["World Bank", "WHO", "Our World in Data", "ONU Dicc", "ISO-alpha3 code"]]
countries
```

```python
countries_df = countries.rename(columns={
                       "World Bank":"world_bank", 
                       "WHO":"who", 
                       "Our World in Data":"owid",
                       "ONU Dicc": "onu", 
                       "ISO-alpha3 code":"iso_country"})
countries_df
```

```python
countries_df.to_csv("clean/extract_countries_all.csv", index=False)
```
