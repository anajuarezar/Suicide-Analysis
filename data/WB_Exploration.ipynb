```python
## Dependencies
import os
import pandas as pd
```

```python
## Set input file
birth = 'wb_birth_rate_per_1k.csv'
path_birth = os.path.join('input', birth)

health = 'wb_current_health_expenditure_per_capita.csv'
path_health = os.path.join('input', health)

death = 'wb_death_rate_per_1k.csv'
path_death = os.path.join('input', death)

gdp = 'wb_gdp_per_capita.csv'
path_gdp = os.path.join('input', gdp)

health_pct = 'wb_health_expenditure_pct.csv'
path_health_pct = os.path.join('input', health_pct)

income = 'wb_income_per_capita.csv'
path_income = os.path.join('input', income)

population = 'wb_population_total.csv'
path_population = os.path.join('input', population)

suicide = 'wb_suicide_rate_per_100k.csv'
path_suicide = os.path.join('input', suicide)

unemployment = 'wb_unemployment_pct.csv'
path_unemployment = os.path.join('input', unemployment)
```

```python

## Load CSV as DataFrame
birth_df = pd.read_csv(path_birth, skiprows=[1,2,3], header=1)
health_df = pd.read_csv(path_health, skiprows=[1,2,3], header=1)
death_df = pd.read_csv(path_death, skiprows=[1,2,3], header=1)
gdp_df = pd.read_csv(path_gdp, skiprows=[1,2,3], header=1)
healthPct_df = pd.read_csv(path_health_pct, skiprows=[1,2,3], header=1)
income_df = pd.read_csv(path_income, skiprows=[1,2,3], header=1)
population_df = pd.read_csv(path_population, skiprows=[1,2,3], header=1)
suicide_df = pd.read_csv(path_suicide, skiprows=[1,2,3], header=1)
unemployment_df = pd.read_csv(path_unemployment, skiprows=[1,2,3], header=1)
```

```python
## Get countries list
country_list = birth_df['Country Name'].unique()
print('Countries List =', len(country_list))
```

```python
## Set years range
start_year = 1990
end_year = 2020
years_list = [str(i) for i in range(start_year, end_year+1)]
print('Num years =', len(years_list))
print(years_list)
```

```python
## Set initial values
country_col = []
year_col = []
birth_rate = []
health = []
death_rate = []
gdp = []
health_pct = []
income = []
population = []
suicide_rate = []
unemployment = []

## Build columns loop
for country in country_list:
    for year in years_list:
        country_col.append(country)
        year_col.append(year)
     
    row_values = birth_df.loc[birth_df['Country Name'] == country][years_list].values[0]
    health_values = health_df.loc[health_df['Country Name'] == country][years_list].values[0]
    death_values = death_df.loc[death_df['Country Name'] == country][years_list].values[0]
    gdp_values = gdp_df.loc[gdp_df['Country Name'] == country][years_list].values[0]
    pct_values = healthPct_df.loc[healthPct_df['Country Name'] == country][years_list].values[0]
    income_values = income_df.loc[income_df['Country Name'] == country][years_list].values[0]
    population_values = population_df.loc[population_df['Country Name'] == country][years_list].values[0]
    suicide_values = suicide_df.loc[suicide_df['Country Name'] == country][years_list].values[0]
    unemployment_values = unemployment_df.loc[unemployment_df['Country Name'] == country][years_list].values[0]
    
    for x in row_values:
        birth_rate.append(x)
        
    for x in health_values:
        health.append(x)
    
    for x in death_values:
        death_rate.append(x)
    
    for x in gdp_values:
        gdp.append(x)
    
    for x in pct_values:
        health_pct.append(x)
        
    for x in income_values:
        income.append(x)
        
    for x in population_values:
        population.append(x)
        
    for x in suicide_values:
        suicide_rate.append(x)
        
    for x in unemployment_values:
        unemployment.append(x)
```

```python
## Build DataFrame
extract_all = pd.DataFrame({
    'country': country_col,
    'year': year_col,
    'birth_rate': birth_rate,
    'health': health,
    'death_rate': death_rate,
    'gdp': gdp,
    'health_pct': health_pct,
    'income': income,
    'population': population,
    'suicide_rate': suicide_rate,
    'unemployment': unemployment
    
})
```
