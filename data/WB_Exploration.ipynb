```python
## Dependencies
import os
import pandas as pd
```

```python
## Set input file
filename = 'wb_population_total.csv'
population = os.path.join('input', filename)

## Load CSV as DataFrame
population_df = pd.read_csv(population, skiprows=[1,2,3], header=1)
population_df
```

```python
## Get countries list
countries_list = population_df['Country Name'].unique()
# print(countries_list)
```

```python
## Set years range
start_year = 1990
end_year = 2020
years_list = [str(i) for i in range(start_year, end_year+1)]
print(years_list)
```

```python
## Set initial values
country_col = []
year_col = []
data_col = []

## Build columns loop
for country in countries_list:
    for year in years_list:
        country_col.append(country)
        year_col.append(year)
        
    row_values = population_df.loc[population_df['Country Name'] == country][years_list]
    for value in row_values:
        data_col.append(value)
```

```python
## Build DataFrame
population_extract = pd.DataFrame({
    'country': country_col,
    'year': year_col,
    'population': data_col
})
```
